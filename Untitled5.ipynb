# 📦 Import Libraries & Create Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML

data = {
    "Lead": ["A", "B", "C", "D", "E", "F"],
    "Stage": ["Prospect", "Negotiation", "Closed Won", "Prospect", "Closed Lost", "Negotiation"],
    "Value": [10000, 20000, 15000, 12000, 8000, 18000]
}

df = pd.DataFrame(data)
```

```python
display(HTML("""
<div style="background: linear-gradient(90deg, #4CAF50, #2E7D32);
padding: 20px; border-radius: 12px; color: white;">
<h1>🤖 AI Pipeline Strategist Dashboard</h1>
<p>AI-powered sales insights using Gemini</p>
</div>
"""))
```

```python
total_value = df["Value"].sum()
won_value = df[df["Stage"] == "Closed Won"]["Value"].sum()
lost_value = df[df["Stage"] == "Closed Lost"]["Value"].sum()

display(HTML(f"""
<div style="display:flex; gap:20px; margin-top:20px;">
  <div style="flex:1; background:#E3F2FD; padding:20px; border-radius:12px;">
    <h3>💰 Total Pipeline</h3>
    <h2>{total_value}</h2>
  </div>

  <div style="flex:1; background:#E8F5E9; padding:20px; border-radius:12px;">
    <h3>✅ Won Deals</h3>
    <h2>{won_value}</h2>
  </div>

  <div style="flex:1; background:#FFEBEE; padding:20px; border-radius:12px;">
    <h3>❌ Lost Deals</h3>
    <h2>{lost_value}</h2>
  </div>
</div>
"""))
```

```python
stage_counts = df["Stage"].value_counts()

plt.figure()
stage_counts.plot(kind="bar")
plt.title("Pipeline Stage Distribution")
plt.xlabel("Stage")
plt.ylabel("Count")
plt.show()
```
